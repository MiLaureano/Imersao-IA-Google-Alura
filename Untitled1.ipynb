```python
DOCUMENT1 = {
    "Título": "Categorias de base definem campeoes",
    "Conteúdo": "As categorias de base definiram nesta sexta-feira (24 de junho) os campeões nacionais da prova de resistência. O Campeonato Brasileiro de Ciclismo de Estrada Elite, Sub23 e Base está sendo disputado por mais de 450 ciclistas na cidade de Palmas, no Tocantins."}

DOCUMENT2 = {
    "Título": "Ciclismo de Estrada do Brasil participa do Tour da Colômbia com duas equipes",
    "Conteúdo": " A Seleção Brasileira e a Swift Carbon participam da competição que vai acontecer entre 6 e 11 de fevereiro de 2024. O evento é classificado como 2.1 na UCI, o que significa que uma parte das equipes participantes fazem parte do World Tour e os outros times serão das classes Continental, Pro Continental e Seleções Nacionais. A Seleção Brasileira será composta pelos atletas Nicolas Sessler, André Eduardo Gohr, Euller Magno, Victor Paula, Werik Kaua e Andrey Braguini. Além disso, a equipe brasileira Swift Carbon Pro Cycling também marcará presença no evento, assim como Vinicius Rangel, que defende a equipe profissional Movistar Team."}

DOCUMENT3 = {
    "Título": "Brasileiras brilham no Pan-Americano de Ciclismo de Estrada",
    "Conteúdo": "Apesar de voltarmos pra casa com apenas um pódio, o bronze de Mayra Costa, o Pan-Americano de Ciclismo de Estrada 2023, que aconteceu na Cidade do Panamá entre 18 e 23 de abril, rendeu ao Brasil alguns bons resultados, especialmente no que se refere às mulheres. O país sai do Pan com a certeza de um belo futuro no ciclismo feminino, diz Marcelo Vanderlinde, da página País do Ciclismo no Instagram e Twitter. E, mais importante, ao que tudo indica, a quinta colocação da Ana Vitória Magalhães (a Tota) na elite feminina pode significar uma vaga nos Jogos Olímpicos de Paris em 2024."}

documents = [DOCUMENT1, DOCUMENT2, DOCUMENT3]
```

```python
df = pd.DataFrame(documents)
df.columns = ["Titulo", "Conteudo"]
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Titulo"], row["Conteudo"]), axis=1)
df
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]

  produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]["Conteudo"]
```

```python
consulta = "quais os atletas da seleção brasileira?"

trecho = gerar_e_buscar_consulta(consulta, df, model)
print(trecho)
```

```python
generation_config = {
  "temperature": 0,
  "candidate_count": 1
}
```

```python
prompt = f"Reescreva esse texto de uma forma mais descontraída, sem adicionar informações que não façam parte do texto: {trecho}"

model_2 = genai.GenerativeModel("gemini-1.0-pro",
                                generation_config=generation_config)
response = model_2.generate_content(prompt)
print(response.text)
```
